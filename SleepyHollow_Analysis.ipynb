```python
# Import all the goods
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
plt.style.use('seaborn-whitegrid')

# Import the data files
xl_01 = pd.ExcelFile('SleepyHollow_Frequencies.xlsx')
all_data = xl_01.parse("Data")
all_data
```

```python
periods = all_data[['Paragraph Number', 'Word Count', 'Periods', 'Periods per word']]
periods
```

```python
central_line = sum(periods['Periods per word'])/periods['Periods per word'].count()
central_line
```

```python
upper_limit = central_line + 3*np.sqrt( (central_line * (1 - central_line)) / periods['Periods per word'])
upper_limit
```

```python
lower_limit = central_line - 3*np.sqrt( (central_line * (1 - central_line)) / periods['Periods per word'])
lower_limit
```
