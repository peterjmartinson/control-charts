```python
# Import all the goods
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
plt.style.use('seaborn-whitegrid')

# Import the data files
xl_01 = pd.ExcelFile('SleepyHollowPData.xlsx')
semicolons = xl_01.parse("Semicolons")
dashes = xl_01.parse("Dashes")
```

```python
semicolon_central_line = semicolons['Central Line'].max()
semicolon_upper_limit = semicolons['Upper Limit']
semicolon_lower_limit = semicolons['Lower Limit']

dash_central_line = dashs['Central Line'].max()
dash_upper_limit = dashs['Upper Limit']
dash_lower_limit = dashs['Lower Limit']

```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, sharex='col')

ax1.step(semicolon_lower_limit.index, semicolon_lower_limit.values, where='mid')
ax1.step(semicolon_upper_limit.index, semicolon_upper_limit.values, where='mid')
ax1.axhline(semicolon_central_line)
ax1.plot(semicolons['x/n'])
ax1.set_ylabel("Semicolons per Sentence")

ax2.step(dash_lower_limit.index, dash_lower_limit.values, where='mid')
ax2.step(dash_upper_limit.index, dash_upper_limit.values, where='mid')
ax2.axhline(dash_central_line)
ax2.plot(dashes['x/n'])
ax2.set_xlabel("Paragraph Number")
ax2.set_ylabel("Dashes per Sentence")

```

```python
fig, ax = plt.subplots(1, 1, figsize=(20, 10))

SMALL_SIZE = 12
MEDIUM_SIZE = 14
BIGGER_SIZE = 18

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

ax.step(semicolon_lower_limit.index, semicolon_lower_limit.values, where='mid')
ax.step(semicolon_upper_limit.index, semicolon_upper_limit.values, where='mid')
ax.axhline(semicolon_central_line)
ax.plot(semicolons['x/n'], marker='o')
ax.set(xlabel='Paragraph Number', ylabel='Semicolons per Sentence',
       title='Semicolons in The Legend of Sleepy Hollow')
fig.savefig("semicolons.jpg")
```
