```python
##### fig, ax = plt.subplots(1, 1, figsize=(20, 10))

SMALL_SIZE = 12
MEDIUM_SIZE = 14
BIGGER_SIZE = 18

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title

ax.step(semicolon_df['lower limit'].index, semicolon_df['lower limit'].values, where='mid')
ax.step(semicolon_df['upper limit'].index, semicolon_df['upper limit'].values, where='mid')
ax.axhline(chart.getCenterLine())
ax.plot(semicolon_df['data'], marker='o')
ax.set(xlabel='Paragraph Number', ylabel='Semicolons per Sentence',
       title='Semicolons in The Legend of Sleepy Hollow')
#fig.savefig("semicolons.jpg")
```
